```python
# IF ELIF ELSE

score = int(input("Enter the score :"))
if score < 35 :
  print("D grade")
elif score >=35 and score < 60 :
  print("C grade")
elif score >=60 and score < 80 :
  print("B grade")
else :
  print("A grade")


# FOR Loop
for i in range(5) :
  print(i)

print("\n\n")
for i in range(1, 10) :
  print(i)


print("\n\n")
for i in range(1, 10, 3) :
  print(i)

print("\n\n")
for i in range(10, 1, -2) :
  print(i)

print("\n\n")
count = 0
for i in range(1,10) :
  if i % 2 == 0 :
    print(i)
    count += 1
print("total even number =", count)


print("\n\n")
fruits = ["apple", "banana", "cherry"]
for i in  fruits :
  print(i)


# While Loop
print("\n\n")
i=1
while i < 5 :
  print (i)
  i += 1

print("\n\n")
while True :
  name = input("Enter the name :")
  if name == "Done":
    break
  print("Welcome " + name)
  print("Enter \"Done\" as name to Exit")


# Nested Loops
print("\n\n")
color = ["Blue","white","yellow"]
item = ["shirt", "shoe"]
for i in color :
  for j in item :
    print(i,j)

print("\n\n")
for i in range(5,10) :
  for j in range(1,11) :
    print(i,'*',j,"=",i*j)
  print("\n")



```
