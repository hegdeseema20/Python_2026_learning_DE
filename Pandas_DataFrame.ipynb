# Working with multiple DF

```python
import pandas as pd

# -----------------------
# Customers DataFrame
# -----------------------
customers = pd.DataFrame({
    "customer_id": [1, 2, 3, 4, 5, 6],
    "name": ["Alice", "Bob", "Charlie", "David", "Eva", "Frank"],
    "city": ["Vancouver", "Toronto", "Vancouver", "Calgary", "Toronto", "Montreal"],
    "age": [25, 34, 28, 45, 22, 40]
})
print("Customers:\n", customers)
```

```python

# -----------------------
# Products DataFrame
# -----------------------
products = pd.DataFrame({
    "product_id": [101, 102, 103, 104, 105],
    "product_name": ["Laptop", "Phone", "Tablet", "Headphones", "Monitor"],
    "category": ["Electronics", "Electronics", "Electronics", "Accessories", "Electronics"],
    "price": [1200, 800, 400, 150, 300]
})

print("\nProducts:\n", products)
```

```python
# -----------------------
# Orders DataFrame
# -----------------------
orders = pd.DataFrame({
    "order_id": [1001, 1002, 1003, 1004, 1005, 1006, 1007],
    "customer_id": [1, 2, 1, 3, 5, 2, 7],  # NOTE: 7 does not exist in customers
    "product_id": [101, 102, 103, 104, 105, 101, 102],
    "quantity": [1, 2, 1, 3, 2, 1, 1],
    "order_date": pd.to_datetime([
        "2024-01-10", "2024-01-11", "2024-01-15",
        "2024-02-01", "2024-02-10", "2024-03-05", "2024-03-10"
    ])
})

print("\nOrders:\n", orders)
```

```python
orders_customers = pd.merge(orders, customers, on="customer_id", how="inner")
orders_customers
```

```python
orders_customers = pd.merge(orders, customers, on="customer_id", how="left")
orders_customers
```

```python
orders_customers = pd.merge(orders, customers, on="customer_id", how="right")
orders_customers
```

```python
# adding more customers

more_customers = pd.DataFrame({
    "customer_id": [7, 8],
    "name": ["Grace", "Henry"],
    "city": ["Ottawa", "Vancouver"],
    "age": [30, 27]
})

all_customers = pd.concat([customers, more_customers])
all_customers
```

```python
common_ids = set(customers["customer_id"]).intersection(set(orders["customer_id"]))
common_ids
```

```python
df = pd.merge(orders, products, on="product_id", how="left")
df["total_price"] = df["price"] * df["quantity"]
df
```

```python

customer_spending = df.groupby("customer_id")["total_price"].sum()
customer_spending_df = customer_spending.reset_index()
customer_spending_df.sort_values(by = 'total_price' , ascending=False)

#customer_spending_df = df.groupby("customer_id", as_index=False)["total_price"].sum()
# this also works directly as DataFrame
```

finding customer details who had highest total purchase

```python
#as_index=False - to convert the query result into dataframe
df1 = df.groupby("customer_id", as_index=False)["total_price"].sum().sort_values(by ='total_price', ascending=False)
df1
```

```python
top_customer_id = df1.nlargest(1, 'total_price')["customer_id"].iloc[0]

result = all_customers[all_customers["customer_id"] == top_customer_id]
result
```
