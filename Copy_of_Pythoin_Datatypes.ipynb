**📚 PYTHON DATA TYPES & COLLECTIONS**

**1. INTEGER (int)**


```python
age = 35

print("=== Integer (int) ===")
print("Age:", age)
print("Adding five years to age:", age + 5)
print("Double the age:", age * 2)
```

**2. FLOAT (float)**


```python
height = 5.9
weight = 70.5

print("\n=== Float (float) ===")
print("Height after adding 1.1:", height + 1.1)
print("Weight divided by 2:", weight / 2)
print("Height multiplied by 2:", height * 2)
```

**3. STRING (str)**


```python
name = "alice"
city = "vancouver"

print("\n=== String (str) ===")
print("Name capitalized:", name.capitalize())
print("City in uppercase:", city.upper())
print("Concatenated string:", name + " lives here")
print("Number of letters in city:", len(city))
```

**4. BOOLEAN (bool)**


```python
is_student = True
is_adult = False

print("\n=== Boolean (bool) ===")
print("AND:", is_student and is_adult)
print("OR:", is_student or is_adult)
print("NOT:", not is_student)
print("Age < 30?:", age < 30)
```

**5. LIST (list)**


```python
fruits = ["Apple", "Banana", "Cherry"]

print("\n=== List ===")
print("Original:", fruits)
```

```python
# Add
fruits.append("Orange")
print("After append:", fruits)
```

```python
# Modify
fruits[1] = "Blueberry"
print("After modify:", fruits)
```

```python
# Remove
fruits.remove("Cherry")
print("After remove:", fruits)
```

```python
# Insert
fruits.insert(2, "Dragon Fruit")
print("After insert:", fruits)
```

```python
# Pop
fruits.pop()
print("After pop:", fruits)
```

```python
# Extend
fruits.extend(["Mango", "Watermelon"])
print("After extend:", fruits)
```

```python
# Delete slice
del fruits[1:3]
print("After delete slice:", fruits)
```

```python
# Add more for sorting
fruits.extend(["lemon", "dates", "lemon", "grapes"])
print("Before sorting:", fruits)
```

```python
# Sort
fruits.sort()
print("Sorted:", fruits)
```

```python
# Reverse sort
fruits.sort(reverse=True)
print("Reverse sorted:", fruits)
```

```python
# Sort by length
fruits.sort(key=len)
print("Sort by length:", fruits)
```

```python
# Clear
fruits.clear()
print("After clear:", fruits)
```

**6. TUPLE (tuple)**

```python
point = (10, 20, 20)

print("\n=== Tuple ===")
print("Tuple:", point)
print("Access element:", point[1])
print("Count of 20:", point.count(20))

# Immutable (cannot change)
# point[1] = 10  # ❌ This will cause an error
```

**7. SET (set)**


```python
numbers = {1, 6, 3, 8, 2, 7, 2}

print("\n=== Set ===")
print("Original (duplicates removed):", numbers)

numbers.add(9)
print("After adding:", numbers)
```

**8. DICTIONARY (dict)**


```python
student = {
    "name": "Alex",
    "age": 16,
    "grade": "A"
}

print("\n=== Dictionary ===")
print("Original:", student)
```

```python
# Modify
student["age"] = 20
print("After updating age:", student)
```

```python
# Add new key
student["address"] = "King George Blvd"
print("After adding address:", student)
```
