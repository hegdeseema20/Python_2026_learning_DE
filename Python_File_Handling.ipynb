### **File Handling**


"r" → read (default)

"w" → write (overwrites file)

"a" → append (adds to file)

"x" → create (fails if file exists)

"b" → binary mode (e.g., "rb")

"t" → text mode (default)

Checking if File Exists

reading the file if exists

```python
import os

if os.path.exists("test_python.txt"):
    print("File exists\n")

    file = open("test_python.txt","r")
    content = file.read()
    print(content)
    file.close()
    print("file closed ?",file.closed)

else:
    print("File not found")

# You should close files because it:

# Frees system resources
# Prevents file locking issues
# Ensures data is safely written
# Keeps your program stable

# 👉 Best practice: always use with open(...)
```

Instead of manually closing files, use with:

✔ Automatically closes the file
✔ Cleaner and safer

### **CSV Files**

csv.reader → simple lists

csv.DictReader → dictionaries (recommended)

csv.writer → write rows

csv.DictWriter → write dictionaries

```python
# Reading the csv file
import csv
with open("/content/drive/MyDrive/sample_csv_file.csv","r") as f :
  data = csv.reader(f,delimiter=",")
  for row in data :
    print(row)
```

```python
# Reading the csv file by skipping the header
import csv
with open ("/content/drive/MyDrive/sample_csv_file.csv","r") as f :
  data = csv.reader(f,delimiter=",")
  next(data)
  for row in data :
    print(row)
```

```python
# Using DictReader to read csv file - easy to understand - Better way
import csv
with open ("/content/drive/MyDrive/sample_csv_file.csv","r") as f :
  data = csv.DictReader(f,delimiter=",")
  for row in data :
    print(row)
    print(row["name"], row["age"])
```

```python
# Writing into csv file
# Always use newline="" when writing --> Prevents blank lines (especially on Windows)
import csv
with open ("sample_csv_file_w.csv","w", newline = "") as f :
  data_w = csv.writer(f)
  data_w.writerow(["name", "age", "grade"])
  data_w.writerow(["Alice", 20, "A"])
  data_w.writerow(["Bob", 22, "B"])

with open("sample_csv_file_w.csv","r") as f :
  data = csv.reader(f,delimiter=",")
  for row in data :
    print(row)
```

```python
with open("sample_csv_file_w.csv","w", newline = "") as f :
  data_w = csv.writer(f)
  data_w.writerow(["gary", 34, "D"])

with open("sample_csv_file_w.csv","r") as f :
  data = csv.reader(f,delimiter=",")
  for row in data :
    print(row)
```

```python
# Writing into csv file - in append more and multiple rows
rows = [
    ["Charlie", 19, "A"],
    ["David", 21, "C"]
]

with open("sample_csv_file_w.csv","a", newline = "") as f :
  data_w = csv.writer(f)
  data_w.writerows(rows)

with open("sample_csv_file_w.csv","r") as f :
  data = csv.reader(f,delimiter=",")
  for row in data :
    print(row)
```

```python
#Writing with DictWriter

import csv
with open("output.csv", "w", newline = "") as f :
  fieldnames_1 = ["name", "age", "grade"]
  data_w = csv.DictWriter(f, fieldnames=fieldnames_1)

  data_w.writeheader()
  data_w.writerow({"name": "Alice", "age": 20, "grade": "A"})
  data_w.writerow({"name": "Bob", "age": 22, "grade": "B"})

with open("output.csv","r") as f :
  data = csv.DictReader(f,delimiter=",")
  for row in data :
    if row["grade"] == "A" :
      print(row["name"])
```
